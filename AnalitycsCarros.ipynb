<h3 Style= 'color:#f99200'> 4. Manejo de valores faltantes </h3>

<p> Que se hace para el manejo de valores faltantes? </p>

<ol> 
    <li>Identificar como se muestran los valores faltantes</li>
    <li>Contar la cantidad de valores faltantes en el set de datos</li>
    <li>Imputacion de valores numericos</li>
    <li>Imputacion de valores categoricos</li>
    <li>Eliminar filas</li> 
</ol>

<h5 Style= 'color:#f99200'> 4.1 Identificar valores faltantes </h5>

```python
dfAutos.replace('?', np.nan, inplace = True)
dfAutos.head(5)
```

<h5 Style= 'color:#f99200'> 4.2 Contar la cantidad de valores faltantes </h5>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h5 Style= 'color:#f99200'> 4.3 Imputacion de de valores numericos </h5>

```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0)
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
av_horsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
#print('La media es:', av_normalized)
```

```python
dfAutos["normalized"].replace(np.nan, av_normalized, inplace = True)
dfAutos["bore"].replace(np.nan, av_bore, inplace = True)
dfAutos["stroke"].replace(np.nan, av_stroke, inplace = True)
dfAutos["horsepower"].replace(np.nan, av_horsepower, inplace = True)
dfAutos["peak"].replace(np.nan, av_peak, inplace = True)
dfAutos.head(5)
```

<h5 Style= 'color:#f99200'> 4.4 Imputacion de de valores categoricos </h5>

```python
dfAutos['doors'].value_counts().idxmax()
```

```python
dfAutos["doors"].replace(np.nan, 'four', inplace = True)
dfAutos.head(5)
```

<h5 Style= 'color:#f99200'> 4.5 Eliminar filas con valores faltantes </h5>

```python
dfAutos.dropna(subset = ['price'], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
```

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<ol>

<h5 Style= 'color:#f99200'> 5.1 Verificar el Dominio de las Variables</h5>

["normalized"], ["bore"], ["stroke"], ["horsepower"], ["peak"],["price"]

```python
dfAutos[["normalized"]] = dfAutos[["normalized"]].astype("int")
dfAutos.dtypes
```

```python
dfAutos[["horsepower","peak", "price"]] = dfAutos[["horsepower","peak", "price"]].astype("int")
dfAutos.dtypes
```

```python
dfAutos[["bore","stroke"]] = dfAutos[["bore","stroke"]].astype("float")
dfAutos.dtypes
```

```python
dfAutos.describe()
```

<h5 Style= 'color:#f99200'> 5.2 Convertir valores numericos en funcion del medio (Estandarizar) </h5>

```python
dfAutos. head(4)
```

```python
dfAutos["City L/100Km (Calc)"] = 235/dfAutos["City"]
dfAutos["Highway L/100Km (Calc)"] = 235/dfAutos["Higway"]
dfAutos. head(4)
```

<h5 Style= 'color:#f99200'> 5.3 Normalizar valores entre 0 y 1 </h5>

```python
dfTemp = dfAutos[["City L/100Km (Calc)","Highway L/100Km (Calc)","engineSize"]]
dfTemp.head(5)
```

```python
dfTemp.describe()
```

```python
dfAutos["City L/100Km (Norm)"] = dfAutos["City L/100Km (Calc)"]/ dfAutos["City L/100Km (Calc)"].max()
dfAutos["Highway L/100Km (Norm)"] = dfAutos["Highway L/100Km (Calc)"]/ dfAutos["Highway L/100Km (Calc)"].max()
dfAutos["engineSize (Norm)"] = dfAutos["engineSize"]/ dfAutos["engineSize"].max()

dfTemp = dfAutos[["City L/100Km (Calc)","Highway L/100Km (Calc)","engineSize","City L/100Km (Norm)","Highway L/100Km (Norm)","engineSize (Norm)"]]
dfTemp.head(5)
```

<h5 Style= 'color:#f99200'> 5.3 Crear rangos para agrupar datos</h5>

```python
dfTemp = dfAutos["horsepower"]
dfTemp.describe()
```

Pregunta 1: ¿Cuatos Autos existen según marca?

```python
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending = False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,6.9), dpi=100)

eje_x = gf.index.tolist()
data = gf['symboling'].tolist()

bar_colors = ['#89abb4','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
              '#e5e5e5','#fdd1b6']
ax.barh(eje_x, data, color=bar_colors)

 

#--------------------------------------------------------------------------------------------------------------------#
#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO
title = "Grafico 1: Vehiculos"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "serif")


#FORMATO DEL SUBTITULO
subTitle = "Cantidad de vehiculos agrupados segun fabricante"
ax.set_title(subTitle,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

 
#FORMATO DE LOS EJES
etiqueta = "Cantidad de vehiculos (ABS)"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

 
#CONFIGURACION DE COORDENADAS DE LA ETIQUETA DEL EJE X
ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)

#CONFIGURACION EL TAMANO DEL TEXTO DEL EJE Y
ax.tick_params(axis='both',
               labelsize=10)


#CONFIGURACION LA UBICACION DEL GRAFICO
plt.subplots_adjust(top=0.80)
plt.show()
```
