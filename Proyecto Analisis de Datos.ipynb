<h4 style="color: #89abb4">5.Â¿CuÃ¡l es el tipo de propiedades residenciales que se vende con mayor frecuencia, dÃ³nde y con quÃ© costo promedio?:</h4>

```python
# Filtrar el DataFrame para incluir solo propiedades residenciales
residential_df = df_dataset[df_dataset['PropertyType'] == 'Residential']

# Obtener el tipo de propiedades residenciales mÃ¡s comÃºn
most_common_residential_type = residential_df['ResidentialType'].value_counts().idxmax()

# Filtrar el DataFrame para incluir solo el tipo de propiedad residencial mÃ¡s comÃºn
most_common_residential_df = residential_df[residential_df['ResidentialType'] == most_common_residential_type]
most_common_residential_type
```

```python
# Agrupar por ciudad o pueblo y calcular la cantidad de ventas y el costo promedio
city_sales_info = most_common_residential_df.groupby('Town').agg({'SaleAmount(COLONES)': 'count', 'SaleAmount(COLONES)': 'mean'})
city_sales_info
```

```python
# Crear la figura y el eje para el grÃ¡fico
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# Ordenar los datos en orden descendente y seleccionar las top 10 ciudades
city_sales_info_sorted = city_sales_info.sort_values('SaleAmount(COLONES)', ascending=False)
top_n = 10
top_cities = city_sales_info_sorted.index[:top_n]
data = city_sales_info_sorted.loc[top_cities, 'SaleAmount(COLONES)'].values

# Generar el grÃ¡fico de barras para la cantidad de ventas por ciudad o pueblo
bar_colors = ['#c2c199']
ax.barh(top_cities, data, color=bar_colors)
title = "Grafico 5. Ventas de Propiedades Single Family por Ciudad o Pueblo"
plt.suptitle(title,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=14,
             fontweight="bold",
             fontfamily="serif")

subTitle = "Top 10 ciudades con mÃ¡s ventas"
ax.set_title(subTitle,
             horizontalalignment='left',
             x=0.001,
             y=0.80,
             transform=fig.transFigure,
             fontsize=12,
             fontweight="regular",
             fontfamily="serif")

ax.set_xlabel("Cantidad de Ventas",
              ha='left',
              fontsize=10,
              fontweight="regular",
              fontfamily="serif")
ax.set_ylabel("Ciudad o Pueblo",
              ha='left',
              fontsize=10,
              fontweight="regular",
              fontfamily="serif")
ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)
ax.tick_params(axis='both', labelsize=10)
plt.subplots_adjust(top=0.80)

# Mostrar el grÃ¡fico
plt.show()

```

<h4 style="color: #89abb4">6. Â¿CuÃ¡l es margen promedio de ganancia por cada una de las
propiedades vendidas?:</h4>

```python
# Filtrar y agrupar los datos
property_type_groups = df_dataset.groupby('PropertyType')

# Calcular el margen promedio de ganancia por cada tipo de propiedad
margins = property_type_groups.apply(lambda x: (x['SaleAmount'] - x['AssessedValue']).mean())
margins_sorted = margins.sort_values(ascending=False)
margins
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
margins_sorted.plot(kind='barh', color='#125b52', ax=ax)  # Usar kind='barh' para barras horizontales

# Configurar el formato del grÃ¡fico
title = "Grafico 6. Promedio de ganancias de las propiedades vendidas"
plt.suptitle(title,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=14,
             fontweight="bold",
             fontfamily="serif")

subTitle = "Margen promedio de ganancia"
ax.set_title(subTitle,
             horizontalalignment='left',
             x=0.001,
             y=0.80,
             transform=fig.transFigure,
             fontsize=12,
             fontweight="regular",
             fontfamily="serif")

ax.set_xlabel("Margen Promedio", fontsize=12, fontweight="regular", fontfamily="serif")
ax.set_ylabel("Tipo de Propiedad", fontsize=12, fontweight="regular", fontfamily="serif")

ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.subplots_adjust(top=0.70)

# Mostrar el grÃ¡fico
plt.tight_layout()
plt.show()
```
